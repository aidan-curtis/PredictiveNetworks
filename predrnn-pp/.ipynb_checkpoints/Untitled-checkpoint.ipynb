```python
save_data = {}
filename = "/Users/aidancurtis/Desktop/destination2.mp4"
vid = imageio.get_reader(filename,  'ffmpeg')
save_data['input_raw_data'] = np.zeros((60000,3, 180,320))
for num in range(40000):
    image = vid.get_data(num)
    reformatted_img = np.swapaxes(np.swapaxes(image, 2, 1), 1, 0)
    save_data['input_raw_data'][num, :, :, :] = reformatted_img
    if(num%1000 == 0):
        print(num)
```

```python
save_data['dims'] = [[3, 180, 320]]

save_data['clips'] = np.zeros((2, 2000, 2))
for i in range(2000):
    save_data['clips'][0][i][0] = i*20
    save_data['clips'][0][i][1] = 10
    save_data['clips'][1][i][0] = i*20+10
    save_data['clips'][1][i][1] = 10
np.save('train_tennis_data.npy', save_data)

```
